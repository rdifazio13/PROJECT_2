```python
# Initial imports
from alpaca_trade_api.rest import REST, TimeFrame
import alpaca_trade_api as tradeapi
from dotenv import load_dotenv
from datetime import date
load_dotenv()
import requests
import pandas as pd
import numpy as np
import getpass
import os
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```python
# API Info for fetching data, portfolio, etc. from Alpaca
BASE_URL = "https://paper-api.alpaca.markets"

if os.getenv("APCA_API_KEY_ID") is not None:
    ALPACA_API_KEY = os.getenv("APCA_API_KEY_ID")
else:

    ALPACA_API_KEY = getpass.getpass('Type your alpaca api key here, it is safe from viewing.')

if os.getenv("ALPACA_SECRET_KEY") is not None:
    ALPACA_SECRET_KEY = os.getenv("ALPACA_SECRET_KEY")
else:

    ALPACA_SECRET_KEY = getpass.getpass('Type your secret alpaca api key here, it is safe from viewing.')

# Instantiate REST API Connection - Account
api_account = tradeapi.REST(key_id=ALPACA_API_KEY, secret_key=ALPACA_SECRET_KEY, raw_data=True,base_url=BASE_URL, api_version='v2')

# Instantiate REST API Connection - Trade
api_trade = tradeapi.REST(key_id=ALPACA_API_KEY, secret_key=ALPACA_SECRET_KEY, raw_data=False, base_url=BASE_URL, api_version='v2')
```

```python
# Get the start date and end dates
start = "2009-01-03" # Bitcoin's genesis block date
end = "2022-03-20"
```

```python
df = api_trade.get_crypto_bars("BTCUSD", TimeFrame.Day, start, end, exchanges=['CBSE']).df

df = df.drop(columns="exchange") #drop exchange column so df can run through model
```

```python
df.tail()
```

```python
df["returns"] = np.log(df['close']).diff().dropna()
```

```python
df = df.dropna()
```

```python
display(df.head())
display(df.tail())
```

### Short and Long SMA

```python
#create short and long SMA column using finta library
from finta import TA

short_window = 50
long_window = 200

df["SMA-S"] = TA.SMA(df, short_window)
df["SMA-L"] = TA.SMA(df, long_window)

df = df.dropna()

display(df.head())
display(df.tail())
```

### Institution Returns: MSTR, GBTC and BITO

```python
import yfinance as yf

# Request MSTR historic pricing data via finance.yahoo.com API & calculate retuns
MSTR = yf.Ticker('MSTR').history(period='13y')[['Close']]
MSTR["MSTR returns"] = np.log(MSTR['Close']).diff().dropna()

# Request GBTC historic pricing data via finance.yahoo.com API & calculate retuns
GBTC = yf.Ticker('GBTC').history(period='13y')[['Close']]
GBTC["GBTC returns"] = np.log(GBTC["Close"]).diff().dropna()

# Request BITO historic pricing data via finance.yahoo.com API & calculate retuns
BITO = yf.Ticker('BITO').history(period='13y')[['Close']]
BITO["BITO returns"] = np.log(BITO['Close']).diff().dropna()

# concat all returns into a DF 
institutions = pd.concat([MSTR, GBTC, BITO], axis = 1)
institutions = institutions.drop("Close",1)

display(institutions.head())
display(institutions.tail())
```

### Linear Regression

```python
# The independent variable, or the variable used to predict the dependent variable is denoted as x. 
# The dependent variable, or the outcome/output, is denoted as y

# indicate Features (X) from the Target (y)

x = df[['vwap','trade_count']] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets

#x = df.drop(['returns'], axis = 1) for everything except BTC returns

Y = df['returns'] # outcome id like to predict
```

```python
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, Y)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# prediction with sklearn
New_vwap = 41817.138905
New_trade_count = 156823
print ('Predicted returns: \n', regr.predict([[New_vwap, New_trade_count]]))
```

```python
# with statsmodels
X = sm.add_constant(x) # adding a constant
 
model = sm.OLS(Y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```
