```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
# Request MSTR historic pricing data via finance.yahoo.com API
MSTR = yf.Ticker('MSTR').history(period='13y')[['Close']]
MSTR["MSTR returns"] = np.log(MSTR['Close']).diff().dropna()

# View our data
display(MSTR.head())
display(MSTR.tail())
```

```python
# Request GBTC historic pricing data via finance.yahoo.com API
GBTC = yf.Ticker('GBTC').history(period='13y')[['Close']]
GBTC["GBTC returns"] = np.log(GBTC["Close"]).diff().dropna()

# View our data
display(GBTC.head())
display(GBTC.tail())
```

```python
# Request BITO historic pricing data via finance.yahoo.com API
BITO = yf.Ticker('BITO').history(period='13y')[['Close']]
BITO["BITO returns"] = np.log(BITO['Close']).diff().dropna()

# View our data
display(BITO.head())
display(BITO.tail())
```

```python
# merge all returns into a DF 
institutions = pd.concat([MSTR, GBTC, BITO], axis = 1)
institutions = institutions.drop("Close",1)

display(institutions.head())
display(institutions.tail())
```
